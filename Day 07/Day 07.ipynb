```python
import pandas as pd

df = pd.read_csv('../Data/Day 07.txt', header=None, names = ['input'])
df[['hand','bid']] = df['input'].str.split(n=2, expand = True)
df['bid'] = pd.Series(list(map(int, df['bid'])))
df
```

```python
def hand_type(hand):
    
    l = pd.Series(list(hand))
    t = l.value_counts()
    m = max(t)
    
    if m==5: 
        type = 'five of a kind'
    elif m==4: 
        type = 'four of a kind'
    elif m==3:
        if t[1]==2: 
            type = 'full house'
        else: 
            type = 'three of a kind'
    elif m==2:
        if t[1]==2: 
            type = 'two pair'
        else: 
            type = 'one pair'
    else: 
        type = 'high card'

    return(type)
        
```

```python
df['type'] = df.apply(lambda row: hand_type(row['hand']), axis=1)
df
```

```python
# Sort by hand type
sort_order_type = ['five of a kind', 'four of a kind', 'full house', 'three of a kind', 'two pair', 'one pair', 'high card']
df["type"] = pd.Categorical(df["type"], categories = sort_order_type)
# df = df.sort_values(by = "type_c")
# df.reset_index(inplace=True)
df
```

```python
# Get rank based on high card
sort_order_card = ['A', 'K', 'Q', 'J', 'T', '9', '8', '7', '6', '5', '4', '3', '2']
alphabet = ''.join(sort_order_card)
hand = pd.Series(sorted(df['hand'], key=lambda word: [alphabet.index(c) for c in word]))
df_index = pd.DataFrame({'hand':hand, 'rank_highcard':hand.index})


df = df.merge(df_index, on = 'hand')
df = df.sort_values(['type','rank_highcard'], ascending=False)
df.reset_index(inplace=True)
df['rank'] = df.index + 1
df['product'] = df.apply(lambda row: row['bid']*row['rank'], axis=1)
df
```
